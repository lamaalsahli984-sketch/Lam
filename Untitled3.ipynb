```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns # import seaborn
# set the seaborn style
sns.set_theme (style="whitegrid" )
```

```python
# this defines the step size h for the numerical computation.
h = 0.1
# this creates a grid x from 0 to 2 with intervals of h.
x = np.arange(0, 2*np.pi, h)
# this computes the sin function values for each point in the grid x.
y = np.sin(x)
# this calculates the forward difference of y ’ vector ’ divided byh, which is an ap pr ox ima ti on of the derivative of sin (x).
forward_diff = np . diff ( y)/h
# This defines a new grid x_diff that corresponds to the forward differences . It excludes the last point of x since np. diff reduces the array size by one .
x_diff = x[:-1:]
# This calculates the exact solution for the derivative of cos (x), which is -sin (x) , evaluated at x_diff .
exact_solution = np . cos ( x_diff )
# This code block creates a plot showing both the finite difference
plt . figure ( figsize = ( 12 , 8))
plt . plot ( x_diff , forward_diff , "--", \
label = "Finite difference approximation" )
plt . plot ( x_diff , exact_solution , \
label = "Exact solution" )
plt . legend ()
plt . show ()
# Finally , this computes the maximum error between the numerical derivative and the exact solution, then prints the result .
max_error = max ( abs ( exact_solution - forward_diff ))
print ( max_error )
```
